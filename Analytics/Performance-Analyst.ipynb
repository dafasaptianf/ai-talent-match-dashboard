```python
# === 1️⃣ Siapkan Dataframe ===
df = df_analysis_valid.copy()

# Kolom kompetensi (pastikan ada)
pillar_codes = ['GDR', 'CEX', 'IDS', 'QDD', 'STO', 'SEA', 'VCU', 'LIE', 'FTC', 'CSI']
available_pillars = [p for p in pillar_codes if p in df.columns]

# Filter hanya rating = 5
df_high = df[df['rating'] == 5].copy()
if df_high.empty:
    raise ValueError("Tidak ada karyawan dengan rating 5 ditemukan!")

# === 2️⃣ Hirarki Analisis ===
hierarchy_cols = ['company', 'area', 'directorate', 'division', 'department', 'position']

results = []

# === 3️⃣ Looping Multi-level ===
for company in df_high['company'].dropna().unique():
    df_c = df_high[df_high['company'] == company]
    
    for area in df_c['area'].dropna().unique():
        df_a = df_c[df_c['area'] == area]
        
        for directorate in df_a['directorate'].dropna().unique():
            df_d = df_a[df_a['directorate'] == directorate]
            
            for division in df_d['division'].dropna().unique():
                df_v = df_d[df_d['division'] == division]
                
                for department in df_v['department'].dropna().unique():
                    df_dep = df_v[df_v['department'] == department]
                    
                    for position in df_dep['position'].dropna().unique():
                        df_pos = df_dep[df_dep['position'] == position]
                        
                        if df_pos.empty:
                            continue
                        
                        # Hitung rata-rata pilar
                        mean_scores = df_pos[available_pillars].mean()
                        if mean_scores.isna().all():
                            continue
                        
                        # Ambil 2 pilar tertinggi
                        top2 = mean_scores.sort_values(ascending=False).head(2)
                        
                        results.append({
                            "company": company,
                            "area": area,
                            "directorate": directorate,
                            "division": division,
                            "department": department,
                            "position": position,
                            "top1_pillar": top2.index[0],
                            "top1_score": round(top2.iloc[0], 2),
                            "top2_pillar": top2.index[1],
                            "top2_score": round(top2.iloc[1], 2),
                            "n_employees": len(df_pos)
                        })

# === 4️⃣ Simpan ke DataFrame ===
df_success_formula = pd.DataFrame(results)
df_success_formula = df_success_formula.sort_values(
    ["company", "area", "directorate", "division", "department", "position"]
).reset_index(drop=True)

print("✅ Success Formula extracted per hierarchical level.")
print(df_success_formula.head())

# === 5️⃣ Simpan ke CSV ===
os.makedirs("output", exist_ok=True)
df_success_formula.to_csv("output/success_formula_hierarchy.csv", index=False)
print("\n💾 Saved to 'output/success_formula_hierarchy.csv'")
```

```python
def plot_top_pillars(df, group_col, top_n=10):
    """
    Visualisasi pilar dominan berdasarkan level organisasi.
    group_col: nama kolom seperti 'company', 'area', 'directorate', dst.
    top_n: berapa banyak pilar teratas yang akan ditampilkan.
    """
    df_melt = df.melt(
        id_vars=[group_col],
        value_vars=['top1_pillar', 'top2_pillar'],
        var_name='pillar_rank',
        value_name='pillar'
    )

    # Hitung frekuensi tiap pilar per level
    df_freq = (
        df_melt.groupby([group_col, 'pillar'])
        .size()
        .reset_index(name='count')
        .sort_values(['count'], ascending=False)
    )

    unique_groups = df_freq[group_col].unique()

    for group in unique_groups:
        df_g = df_freq[df_freq[group_col] == group].nlargest(top_n, 'count')

        plt.figure(figsize=(8, 4))
        sns.barplot(
            data=df_g,
            x='count', y='pillar',
            palette='Blues_r'
        )
        plt.title(f"Top {top_n} Dominant Competency Pillars in {group_col.title()}: {group}")
        plt.xlabel("Frequency of Appearance (Top1 & Top2)")
        plt.ylabel("Competency Pillar")
        plt.tight_layout()
        plt.show()
```

```python
# 5️⃣ Dominan per DEPARTMENT
plot_top_pillars(df_success_formula, 'department', top_n=8)
```

```python
df_profiles_psych['mbti'] = df_profiles_psych['mbti'].str.strip().str.upper()

typo_map = {
    'INFTJ': 'INFJ',  # Memperbaiki typo 'INFTJ'
}

df_profiles_psych['mbti'] = df_profiles_psych['mbti'].replace(typo_map)
df_profiles_psych['mbti'].unique()
```

```python
# --- ASUMSI ---
# df_analysis_valid: DataFrame 'long' (5 baris/karyawan, 2021-2025)
# df_profiles_psych: DataFrame psikometri (1 baris/karyawan)

print("\n--- Menjalankan Analisis 2 (Diperluas): Evolusi Profil NUMERIK ---")

years_to_analyze = [2021, 2022, 2023, 2024, 2025]
numeric_cols_to_analyze = ['pauli', 'faxtor', 'iq', 'gtq', 'tiki'] 
all_cohort_stats = [] 

# --- 1. Buat Kohort (Grup) per Tahun ---
for year in years_to_analyze:
    ids_this_year = df_analysis_valid[
        (df_analysis_valid['year'] == year) & 
        (df_analysis_valid['rating'] == 5)
    ]['employee_id'].unique()
    
    profile_cohort = df_profiles_psych[
        df_profiles_psych['employee_id'].isin(ids_this_year)
    ]
    
    if not profile_cohort.empty:
        medians = profile_cohort[numeric_cols_to_analyze].median()
        medians['year'] = year 
        all_cohort_stats.append(medians)

# --- 2. Tampilkan Hasil Analisis Evolusi ---
stats_df = pd.DataFrame(all_cohort_stats).set_index('year')
print("\n--- Hasil Analisis Evolusi (Median per Kohort) ---")
print(stats_df)

# --- 3. CONTOH VISUALISASI (Faceted Line Plot) ---
# 'Melt' data agar ramah-seaborn
stats_df_long = stats_df.reset_index().melt(id_vars='year', var_name='Metric', value_name='Median Score')

# Ubah 'year' menjadi string agar diperlakukan sebagai kategori diskrit
stats_df_long['year'] = stats_df_long['year'].astype(str)

print("\nMembuat Faceted Line Plot dengan label median di setiap titik...")

# --- FUNGSI BARU UNTUK MENAMBAHKAN LABEL ---
# Fungsi ini akan dipanggil untuk setiap plot di dalam grid
def add_labels(x, y, **kwargs):
    ax = plt.gca() # Dapatkan axes (plot) saat ini
    # Iterasi melalui setiap titik data (x, y)
    for i in range(len(x)):
        ax.text(
            x.iloc[i],                  # Koordinat x (tahun)
            y.iloc[i],                  # Koordinat y (skor)
            f"{y.iloc[i]:.2f}",       # Teks label (format 2 desimal)
            ha='center',                # Penjajaran horizontal
            va='bottom',                # Penjajaran vertikal (sedikit di atas titik)
            fontsize=8,                 # Ukuran font
            color='black'               # Warna font
        )

# --- MEMBUAT PLOT ---
# Menambahkan 'sharex=False' untuk memaksa semua plot menampilkan label X
g = sns.FacetGrid(stats_df_long, col="Metric", col_wrap=3, height=3, aspect=1.5, 
                  sharey=False, 
                  sharex=False) 

# LAPISAN 1: Gambar garis plot
g.map(sns.lineplot, "year", "Median Score", marker='o', sort=False) 

# --- BARIS BARU: LAPISAN 2 ---
# Tambahkan label teks ke setiap plot
g.map(add_labels, "year", "Median Score")

# Atur judul dan label
g.set_titles("{col_name}")
g.set_axis_labels("Tahun", "Median Skor")
plt.tight_layout()
plt.show()
```

```python
years_to_analyze = [2021, 2022, 2023, 2024, 2025]
cohort_profiles = {} # Dictionary untuk menyimpan profil tiap kohort
cohort_stats = {}    # Dictionary untuk menyimpan statistik IQ

# --- 1. Buat Kohort (Grup) per Tahun ---
for year in years_to_analyze:
    
    # 1a. Dapatkan ID karyawan yang 'rating 5' PADA TAHUN itu
    ids_this_year = df_analysis[
        (df_analysis['year'] == year) & 
        (df_analysis['rating'] == 5)
    ]['employee_id'].unique()
    
    print(f"Tahun {year}: Ditemukan {len(ids_this_year)} high performers.")
    
    # 1b. Ambil profil statis (dari df_profiles_psych) HANYA untuk ID tersebut
    profile_cohort = df_profiles_psych[
        df_profiles_psych['employee_id'].isin(ids_this_year)
    ]
    
    # 1c. Simpan profil kohort ini
    cohort_profiles[year] = profile_cohort
    
    # 1d. (Contoh) Hitung dan simpan median IQ untuk kohort ini
    cohort_stats[year] = profile_cohort['iq'].median()
```

```python
# --- ASUMSI (Lanjutan dari kode di atas) ---
# Kita masih memiliki 'cohort_profiles', yaitu dictionary berisi:
# cohort_profiles[2021] = DataFrame profil high-performer 2021
# cohort_profiles[2022] = DataFrame profil high-performer 2022
# ...dst.

print("\n--- Menjalankan Analisis 2 (Diperluas): Evolusi Profil KATEGORIKAL ---")

# --- 1. Analisis untuk MBTI ---
mbti_distribution = {}
for year in years_to_analyze:
    cohort_df = cohort_profiles.get(year) # Ambil profil kohort
    if cohort_df is not None and not cohort_df.empty:
        # Hitung persentase distribusi (normalize=True)
        mbti_distribution[year] = cohort_df['mbti'].value_counts(normalize=True)

# Ubah dictionary hasil menjadi DataFrame
mbti_heatmap_data = pd.DataFrame(mbti_distribution).fillna(0) * 100 # Kali 100 untuk persentase
print("\nDistribusi Persentase MBTI per Kohort (%):")
print(mbti_heatmap_data)

# --- 2. VISUALISASI (Heatmap) ---
plt.figure(figsize=(10, 8))
sns.heatmap(
    mbti_heatmap_data, 
    annot=True,        # Tampilkan angka persentase
    fmt=".1f",         # Format 1 angka desimal
    cmap="viridis",    # Skema warna
    linewidths=.5
)
plt.title("Evolusi Distribusi Tipe MBTI pada High Performer (2021-2025)")
plt.xlabel("Tahun")
plt.ylabel("Tipe MBTI")
plt.show()


# --- 3. Analisis untuk DISC_WORD (Lakukan hal yang sama) ---
disc_distribution = {}
for year in years_to_analyze:
    cohort_df = cohort_profiles.get(year)
    if cohort_df is not None and not cohort_df.empty:
        disc_distribution[year] = cohort_df['disc_word'].value_counts(normalize=True)

disc_heatmap_data = pd.DataFrame(disc_distribution).fillna(0) * 100
print("\nDistribusi Persentase DISC per Kohort (%):")
print(disc_heatmap_data)

# --- 4. VISUALISASI (Heatmap) ---
plt.figure(figsize=(10, 6)) # Ukuran disesuaikan
sns.heatmap(
    disc_heatmap_data, 
    annot=True, 
    fmt=".1f", 
    cmap="viridis",
    linewidths=.5
)
plt.title("Evolusi Distribusi Tipe DISC pada High Performer (2021-2025)")
plt.xlabel("Tahun")
plt.ylabel("Tipe DISC")
plt.show()
```

```python
# --- ASUMSI ---
# df_analysis: DataFrame 'long' (5 baris/karyawan, 2021-2025)
# df_strengths: DataFrame 'long' (banyak baris/karyawan) dari df_strengths.csv

print("--- Menjalankan Analisis 3 (REVISI V2): Evolusi Profil Strengths (Metode 5 Tema/Karyawan) ---")

years_to_analyze = [2021, 2022, 2023, 2024, 2025]
strengths_distribution = {} 

# --- 1. Bersihkan Data 'Strengths' Terlebih Dahulu ---
# 1a. Bersihkan data 'theme' yang kosong
strengths_cleaned = df_strengths.dropna(subset=['theme'])

# --- PERBAIKAN DI SINI ---
# 1b. Ambil 5 TEMA TERATAS (peringkat terendah) UNTUK SETIAP KARYAWAN
#    Ini menerapkan saran Anda:
#    - Urutkan berdasarkan 'rank'
#    - Ambil 5 baris pertama (head(5)) untuk SETIAP 'employee_id'
strengths_cleaned_top5_per_employee = strengths_cleaned.sort_values(
    by=['employee_id', 'rank']
).groupby('employee_id').head(5)

print(f"Total data Top 5 Strengths (per karyawan) yang akan dianalisis: {len(strengths_cleaned_top5_per_employee)}")


# --- 2. Analisis Kohort per Tahun ---
for year in years_to_analyze:
    
    # 2a. Dapatkan ID karyawan yang 'rating 5' PADA TAHUN itu
    ids_this_year = df_analysis[
        (df_analysis['year'] == year) & 
        (df_analysis['rating'] == 5)
    ]['employee_id'].unique()
    
    # 2b. Ambil profil 'df_strengths' (Top 5/karyawan) HANYA untuk ID tersebut
    cohort_strengths = strengths_cleaned_top5_per_employee[
        strengths_cleaned_top5_per_employee['employee_id'].isin(ids_this_year)
    ]
    
    # 2c. Hitung frekuensi (persentase) tema untuk kohort ini
    if not cohort_strengths.empty:
        # Hitung persentase (normalize=True)
        theme_counts = cohort_strengths['theme'].value_counts(normalize=True)
        strengths_distribution[year] = theme_counts

# --- 3. Ubah Hasil Menjadi DataFrame untuk Heatmap ---
strengths_heatmap_data = pd.DataFrame(strengths_distribution).fillna(0) * 100 

# --- 4. Filter untuk 15 Tema Teratas agar Heatmap Rapi ---
top_15_themes = strengths_heatmap_data.sum(axis=1).nlargest(15).index
strengths_heatmap_data_top15 = strengths_heatmap_data.loc[top_15_themes]

print("\nDistribusi Persentase 15 Strengths Teratas per Kohort (%):")
print(strengths_heatmap_data_top15)

# --- 5. Visualisasikan (Heatmap) ---
plt.figure(figsize=(10, 10))
sns.heatmap(
    strengths_heatmap_data_top15, 
    annot=True,        
    fmt=".1f",         
    cmap="viridis",    
    linewidths=.5
)
plt.title("Evolusi Distribusi 15 'Strengths' Teratas (Top 5 per Karyawan)\npada High Performer (2021-2025)", fontsize=14)
plt.xlabel("Tahun")
plt.ylabel("Tema (Theme)")
plt.show()
```

### Success Formula

Tinjauan komprehensif dari semua visualisasi dan analisa data yang telah dilakukan berikut penjelasannya (rule-based logic)

Analisa saya berdasar kepada Evolusi Kohort per Tahun, yang di mana Pola Sukses di perusahaan ini tidak statis namun terus berevolusi.

Hipotesis Utama: Pola Sukses yang efektif harus beradaptasi dari tahun ke tahun, mencerminkan pergeseran profil high-performer yang telah diamati.


"Success Formula" (Rule-Based Logic)
Seorang karyawan dianggap memiliki Pola Sukses (High Performer) jika memenuhi minimal 3 dari 4 kriteria berikut. Kriteria ini didasarkan pada tren yang paling konsisten yang diamati dari kohort high-performer 2021-2025.

1. Kompetensi Inti (Temuan dari Radar Chart)
Karyawan menunjukkan keunggulan yang jelas di pilar-pilar fundamental yang secara konsisten membedakan high-performer (Rating 5) dari yang lain (Rating 1-4) di semua tahun.

- Skor pilar GDR (Leadership/Direction) >= 4.0 
- Skor pilar CEX (Customer Focus) >= 4.0

2. Profil Perilaku (Temuan dari Heatmap Strengths)
Karyawan memiliki tema bakat (Top 5 per Karyawan) yang secara konsisten paling umum muncul di kohort high-performer, terutama di tahun-tahun terakhir.

- Memiliki 'Achiever' dalam Top 5 Strengths ATAU
- Memiliki 'Learner' dalam Top 5 Strengths.

3. Profil Kognitif & Kepribadian (Temuan dari Line Plot & Heatmap MBTI)
Profil kognitif high-performer sangat berubah-ubah, menunjukkan tidak ada satu "skor IQ" yang pasti. Namun, tipe kepribadian tertentu menunjukkan frekuensi yang lebih tinggi secara konsisten.

- Kognitif: Skor iq >= 105 (Batas bawah median yang paling sering muncul di line plot, menunjukkan standar minimum yang konsisten).

- Kepribadian: Tipe mbti (ENFP) adalah salah satu dari tipe yang paling sering muncul.

4. Konteks & Pengalaman (Temuan dari Box Plot Interaksi)
Pola sukses sangat bergantung pada interaksi antara latar belakang dan pengalaman. Analisis Box Plot menunjukkan bahwa masa kerja adalah pembeda utama, terutama untuk level pendidikan yang lebih tinggi.

- Jika Pendidikan = 'S2', maka Masa Kerja >= 48 bulan.

- Jika Pendidikan = 'S1', maka Masa Kerja >= 36 bulan.

- Jika Pendidikan = 'D3' atau 'SMA', Masa Kerja >= 50 bulan
